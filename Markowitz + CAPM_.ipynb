# Markowitz Portfolio

The Markowitz model, also known as the Modern Portfolio Theory (MPT), was developed by Harry Markowitz and introduced in his paper "Portfolio Selection," published in the "Journal of Finance" in 1952. The model aims to construct an optimal investment portfolio that maximizes returns for a given level of risk, or equivalently, minimizes risk for a given level of expected return. It is based on the premise that investors are risk-averse and wish to optimize their portfolios accordingly.

The Markowitz model involves the selection of a mix of asset classes (such as stocks, bonds, etc.) to achieve diversification, thereby reducing the overall risk of the portfolio. The key concepts in the Markowitz model include expected returns, variance, covariance, and correlation between the returns of different assets.

Here's a brief overview of the formulas used in the Markowitz model:

Expected Portfolio Return
The expected return of a portfolio is calculated as a weighted sum of the individual expected returns of the assets within the portfolio.

image.png

Portfolio Variance
The variance of the portfolio's return, a measure of risk, is not simply the weighted average of the individual variances due to diversification effects. Instead, it also incorporates the covariances or correlations among the asset returns.

image.png


Objective
The objective in the Markowitz model is to find the set of weights
 that minimizes the portfolio's risk for a given level of expected return, or equivalently, maximizes the expected return for a given level of risk. This involves solving an optimization problem, often subject to constraints (e.g., no short selling, budget constraint).

```python
def portfolioreturn(weights):
    return np.dot(df.mean(), weights)
```

```python
def portfoliostd(weights):
    return np.sqrt(np.dot(np.dot(df.cov(), weights),weights.T))
```

```python
def weightscreator(df):
    rand = np.random.random(len(df.columns))
    rand /= rand.sum()
    return rand
```

```python
import plotly.graph_objects as go
import numpy as np


risk_threshold = 0.0085
interval_start = 0.005

fig = go.Figure()

max_returns_indices = []
max_returns_variances = []
max_returns_values = []

# Ensure that the risk does not exceed the threshold
for variance_threshold in np.arange(interval_start, risk_threshold, 0.001):
    mask = (variances >= variance_threshold - 0.001) & (variances <= variance_threshold + 0.001)
    subset_returns = returns[mask]

    if len(subset_returns) > 0:
        max_returns_index = np.where(mask)[0][np.argmax(subset_returns)]
        max_returns_indices.append(max_returns_index)
        max_returns_variances.append(variances[max_returns_index])
        max_returns_values.append(subset_returns[np.argmax(subset_returns)])

fig.add_trace(go.Scatter(x=variances, y=returns, mode='markers', marker=dict(size=5)))

# Highlight points with the max value of y for each period
fig.add_trace(go.Scatter(x=max_returns_variances, y=max_returns_values, mode='markers',
                         marker=dict(size=6, color='red'),
                         text='Max Return for each Variance'))

# Connect the highlighted points with a line
fig.add_trace(go.Scatter(x=max_returns_variances, y=max_returns_values, mode='lines',
                         line=dict(color='blue', width=2),
                         text='Efficient Frontier'))

# Add annotations for the index of the maximum points
for i, index in enumerate(max_returns_indices):
    fig.add_annotation(
        go.layout.Annotation(
            x=max_returns_variances[i],
            y=max_returns_values[i],
            text=f'Max Index: {index}',
            showarrow=True,
            arrowhead=2,
            arrowcolor='black',
            arrowsize=1,
            arrowwidth=2,
            ax=0,
            ay=-40
        )
    )

fig.update_layout(
    title=f'Mean and variance of returns of randomly generated portfolios with risk threshold',
    xaxis=dict(title='Variance'),
    yaxis=dict(title='Return'),
    width=800,
    height=600
)

fig.show()
```

# Capital Asset Pricing Model (CAPM)

The Capital Asset Pricing Model (CAPM) is a cornerstone of modern financial economics. It provides a linear relationship between the expected return of an asset and its systematic risk, measured by beta (\(\beta\)).

## Formula

The CAPM formula is expressed as:

\[E(R_i) = R_f + \beta_i (E(R_m) - R_f)\]

Where:
- \(E(R_i)\) is the expected return on the capital asset,
- \(R_f\) is the risk-free rate,
- \(\beta_i\) is the beta of the security,
- \(E(R_m)\) is the expected return of the market.


## Capital Market Line (CML)

The Capital Market Line (CML) is a tangent line drawn from the point of the risk-free asset to the feasible region for risky assets on the efficient frontier. It represents the optimal portfolio of risky assets combined with risk-free assets.

### CML Formula

The CML is described by the equation:

\[E(R_p) = R_f + \sigma_p \left(\frac{E(R_m) - R_f}{\sigma_m}\right)\]

Where:
- \(E(R_p)\) is the expected return on the portfolio,
- \(R_f\) is the risk-free rate,
- \(\sigma_p\) is the standard deviation of the portfolio's returns (its total risk),
- \(\sigma_m\) is the standard deviation of the market portfolio's returns,
- \(E(R_m) - R_f\) is the market risk premium.

### Interpretation

- The CML represents portfolios that optimally combine the market portfolio with the risk-free asset to achieve the best possible risk-return trade-off.
- All portfolios on the CML are efficient portfolios. Investors select a point on the CML by their risk preference.
- The slope of the CML (\(\frac{E(R_m) - R_f}{\sigma_m}\)) represents the market price of risk for efficient portfolios.

## Security Market Line (SML)

The Security Market Line (SML) is a graphical representation of the Capital Asset Pricing Model (CAPM). It displays the relationship between the expected return of an investment and its beta, illustrating the risk-return tradeoff in the market.

### SML Formula

The formula for the SML is the same as the CAPM formula, representing the expected return for a given risk (\(\beta\)):

\[E(R_i) = R_f + \beta_i (E(R_m) - R_f)\]

In the SML graph, the y-axis represents the expected return of a security, and the x-axis represents its beta. The slope of the SML is determined by the market risk premium (\(E(R_m) - R_f\)), indicating the additional return expected for taking on extra risk.

### Interpretation

- **Below the SML:** Securities below the SML are considered overvalued, as the market expects a lower return for the given level of risk.
- **On the SML:** Securities on the SML are correctly valued, as their expected returns are in line with their risk.
- **Above the SML:** Securities above the SML are undervalued, offering higher expected returns for their risk level, representing potentially better investment opportunities.
